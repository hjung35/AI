```python
import numpy as np

data = np.random.randn(10, 3)
print("Data is, \n", data, end='\n\n')

N = data.shape[0]
print("N is", N, end='\n\n')
    
    # using random.choice for random index to pull out and take the whole row and column of that index from data
    # to centers, same type of 2d array
idx = np.random.choice(N, 4, replace=False)
print("random idx: ", idx, end='\n\n')

centers = data[idx, :]
print("centers are :\n", centers, end='\n\n')
    
    # iteration for niter times
for i in range(3):
    
    print("i th round: ", i, end='\n')
    print("centers.shape: ", centers.shape, end='\n\n')
    print("centers: ", centers, end='\n\n')
    print("centers with np.newaxis: ", centers[:, np.newaxis, :], end='\n\n')
    sq_dists = np.sum((centers[: ,np.newaxis, :] - data)**2, axis=2)
    print("np.sum :", sq_dists)
    closest  = np.argmin(sq_dists, axis=0) 
    for j in range(2):
        centers[j,:] = data[closest==j,:].mean(axis=0)
```
